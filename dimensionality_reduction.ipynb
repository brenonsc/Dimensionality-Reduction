```python
import os
import matplotlib.pyplot as plt
from PIL import Image
```

```python
image = Image.open('images/Lenna.png')
```

```python
grayscale_image = image.convert('L')
grayscale_image.save(os.path.join('images', 'Lenna_gray.jpg'))
```

```python
binary_image = grayscale_image.point(lambda x: 0 if x < 128 else 255, '1')
binary_image.save(os.path.join('images', 'Lenna_binary.jpg'))
```

```python
plt.figure(figsize=(15, 5))

original_image = Image.open('images/Lenna.png')
grayscale_image_path = os.path.join('images', 'Lenna_gray.jpg')
binary_image_path = os.path.join('images', 'Lenna_binary.jpg')

# Imagem original
plt.subplot(1, 3, 1)
plt.imshow(original_image)
plt.title("Original Image")
plt.axis("off")

# Imagem em tons de cinza
plt.subplot(1, 3, 2)
grayscale_image = Image.open(grayscale_image_path)
plt.imshow(grayscale_image, cmap="gray")
plt.title("Grayscale Image")
plt.axis("off")

# Imagem binarizada
plt.subplot(1, 3, 3)
binary_image = Image.open(binary_image_path)
plt.imshow(binary_image, cmap="gray")
plt.title("Binary Image")
plt.axis("off")

plt.show()
```
